# III- Régularité

```python
import nbformat
nb3=nbformat.read('./Analyse_descriptive_Regularite.ipynb', as_version= nbformat.NO_CONVERT)
```


### Comment se comporte le taux de regularite par région ?

```python
exec(nb3.cells[1].source, globals())
exec(nb3.cells[4].source, globals())
```

On observe qu'à chaque novembre survienne des creux de ponctualité des trains. Les commentaires dans la base suggèrent  diversent raisons comme des constructions, des intempéries, etc.
La plupart des régions ont un taux de regularité en moyenne supérieur à 90 % sur la période 2013-2022. La région Bourgogne-Franche-Compte echappe un peu à cette règle vue ses faible taux de regularité tous en dessous de 90% avant 2019.
### Quelles sont les corrélations qui émergent entre variables quantitatives de la base ?

```python
exec(nb3.cells[7].source, globals())
```

Il existe une corrélation évidente entre le nombre de trains programmés et le nombre de trains ayant effectivement circulé, qui est presque parfaite. Cependant, il y a une corrélation intéressante entre le nombre de trains annulés et le nombre de trains en retard dans le même mois. Cela suggère que les retards et les annulations pourraient être affectés de manière similaire par certains phénomènes durant le même mois.

### Voyons de plus près en regardant par région la relation entre le nombre de trains prévus et le nombre de trains ayant effectivement circulé.

```python
exec(nb3.cells[9].source, globals())
```

On remarque que les régions ayant le plus de trains programmés sont le "Grand Est" et "l'Auvergne-Rhône-Alpes" alors que celles ayant le moins de trains programmés sont la Bretagne et l'Occitanie.
L'Auvergne-Rhône-Alpes fait également partie des régions où le nombre de trains annulés est le plus élevé avec aussi la région "Haut de France".
